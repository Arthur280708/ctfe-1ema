```python
n1, n2, n3, n4 = input('insira os numeros da distribuiçao').split()
n1 = int(n1)
n2 = int(n2)
n3 = int(n3)
n4 = int(n4)
```

```python
# Media
MD = (n1 + n2 + n3 + n4)/4
```

```python
# Desvio
d1 = n1 - MD
d2 = n2 - MD
d3 = n3 - MD
d4 = n4 - MD
d1, d2, d2, d3
```

```python
# Somatoria do quadrado dos desvios
soma = d1**2 + d2**2 + d3**2 + d4**2
#variaçao da distribuiçao
variancia = soma / 4
variancia
```

```python
#Desvio padrao
desvio_padrao = variancia**0.5
desvio_padrao
```
