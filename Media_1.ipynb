Calculo da media fiap


```python
# calculo de media de checkpoints
# 3 cps obrigatorias ----> input
# eliminar menor nota
# calcular media das duas maiores notas
```

```python
# 1. Recebe as notas e já as transforma em uma lista de números (floats)
notas = input('Insira as 3 notas dos checkpoints separadas por espaço: ').split()
notas = [float(n) for n in notas]

# 2. Identifica e remove a menor nota
menor_nota = min(notas)
notas.remove(menor_nota)

# 3. Calcula a média das duas que sobraram
MCP1 = sum(notas) / 2

# 4. Exibe o resultado
print("-" * 30)
print(f"Nota descartada: {menor_nota}")
print(f"Notas consideradas: {notas[0]} e {notas[1]}")
print(f"Sua média de Checkpoints é: {MCP1:.2f}")
print("-" * 30)
```

```python
# calculo de sprints
# calcular media das 2 notas
```

```python
# --- Cálculo de Sprints ---

# 1. Entrada das notas das Sprints
Sprint1, Sprint2 = input('Insira as notas das duas Sprints: ').split()

# Converte para float
Sprint1 = float(Sprint1)
Sprint2 = float(Sprint2)

# 2. Calcular a média das 2 notas
MSP1 = (Sprint1 + Sprint2) / 2

# 3. Exibição do resultado
print("-" * 30)
print(f"Média Final das Sprints: {MSP1:.2f}")
print("-" * 30)
```

```python
# calculo de nota global solution (GS)
# Inserir a nota
# caso nao haja nota de GS, o aluno devera inserir sua nota de Sub
```

```python
# --- Cálculo de Global Solution (GS) ---

print("Cálculo da nota de Global Solution")
fez_gs = input("Você realizou a prova de GS? (S/N): ").upper()

if fez_gs == 'S':
    GS1 = float(input("Insira a nota da GS: "))
else:
    print("Reposição necessária.")
    GS1 = float(input("Insira a nota da Sub: "))

# A variável GS agora contém o valor final (seja da prova oficial ou da sub)
print("-" * 30)
print(f"Nota final de GS considerada: {GS1}")
print("-" * 30)
```

```python
# calculo da media semestral 1(MS1)
# MS1 = 0.6*GS1 + 0.2*MSP1 + 0.2*MCP1]
MS1 = (0.6 * GS1) + (0.2 * MSP1) + (0.2 * MCP1)
print(f"MÉDIA SEMESTRAL FINAL (MS1): {MS1:.2f}")
print("="*30)
```

```python
#repetir para segundo semestre
# MS2
```

```python
# --- CÁLCULO DO SEGUNDO SEMESTRE (MS2) ---

# 1. Checkpoints (MCP2)
# Recebe 3 notas, descarta a menor e faz a média das 2 maiores
c1, c2, c3 = input('Insira as 3 notas dos checkpoints do 2º Semestre: ').split()
c1, c2, c3 = float(c1), float(c2), float(c3)
MCP2 = (c1 + c2 + c3 - min(c1, c2, c3)) / 2

# 2. Sprints (MSP2)
# Média simples das duas notas
s1, s2 = input('Insira as 2 notas das sprints do 2º Semestre: ').split()
s1, s2 = float(s1), float(s2)
MSP2 = (s1 + s2) / 2

# 3. Global Solution (GS2)
# Se não fez a GS2, usa a nota da Sub
fez_gs2 = input("Fez a GS do 2º Semestre? (S/N): ").upper()
if fez_gs2 == 'S':
    GS2 = float(input("Nota da GS2: "))
else:
    GS2 = float(input("Nota da Sub do 2º Semestre: "))

# 4. Média Semestral 2 (MS2)
# Mesmos pesos: 60% GS, 20% Sprints, 20% Checkpoints
MS2 = (0.6 * GS2) + (0.2 * MSP2) + (0.2 * MCP2)

# --- RESULTADO DO SEGUNDO SEMESTRE ---
print("\n" + "="*30)
print(f"RESUMO 2º SEMESTRE")
print(f"MCP2: {MCP2:.2f}")
print(f"MSP2: {MSP2:.2f}")
print(f"GS2:  {GS2:.2f}")
print("-" * 30)
print(f"MÉDIA SEMESTRAL 2 (MS2): {MS2:.2f}")
print("="*30)
```

```python
# media final fiap
MF = 0.40*MS1 + 0.60*MS2
```

```python
# --- CÁLCULO DA MÉDIA FINAL FIAP ---

# 1. Aplicando a fórmula de pesos da FIAP
# MS1 vale 40% e MS2 vale 60%
MF = (0.40 * MS1) + (0.60 * MS2)

print("\n" + "="*30)
print(f"MÉDIA ANUAL (MF): {MF:.2f}")

# 2. Verificação de status
if MF >= 6.0:
    print("SITUAÇÃO: APROVADO! Boas férias!")

elif MF < 4.0:
    print("SITUAÇÃO: REPROVADO. Não há direito a exame.")

else:
    # Caso esteja entre 4.0 e 5.9
    print("SITUAÇÃO: EXAME. Você ainda tem uma chance!")

    # 3. Processo de Exame
    EX = float(input("Insira a nota obtida no Exame: "))
    MF2 = (EX + MF) / 2

    print(f"Nova Média após Exame (MF2): {MF2:.2f}")

    if MF2 >= 6.0:
        print("SITUAÇÃO: APROVADO NO EXAME!")
    else:
        print("SITUAÇÃO: REPROVADO NO EXAME.")

print("="*30)
```
